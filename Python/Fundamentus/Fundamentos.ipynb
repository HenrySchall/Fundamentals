### Operadores Lógicos

| Operador | Descrição                                                | Exemplo                                |
|----------|------------------------------------------------------------|-----------------------------------------|
| `and`    | Retorna `True` se ambas as condições forem verdadeiras     | `(5 > 2) and (10 > 5) → True`           |
| `or`     | Retorna `True` se pelo menos uma condição for verdadeira   | `(5 > 2) or (10 < 5) → True`            |
| `not`    | Inverte o valor lógico                                     | `not (5 > 2) → False`                   |


```python
x = 12 
y = 3.5

print (x > 15 and y < 5)
print (x > 5 or y < 10 and x < 7) # primeira condicäo anula o resto
print (not(x > 5 or y < 10 and x < 7))
```

### Estruturas Condicionais

| Estrutura | Uso                                                                 |
|----------|----------------------------------------------------------------------|
| `if`     | Executa um bloco se a condição for verdadeira                        |
| `elif`   | Verifica outra condição se o `if` for falso                           |
| `else`   | Executa se todas as condições anteriores forem falsas                 |
| `for`    | Itera sobre sequências (`list`, `tuple`, `str`, `dict`)               |
| `while`  | Executa enquanto a condição for verdadeira                            |
| `break`  | Interrompe um loop imediatamente                                      |
| `continue` | Pula para a próxima iteração do loop                                |
| `pass`   | Placeholder sem impacto no código                                     |


### Resumo POO

| Conceito     | Descrição                                 | Exemplo                                |
|--------------|---------------------------------------------|-----------------------------------------|
| **Classe**       | Modelo para criar objetos                 | `class Carro:`                           |
| **Objeto**       | Instância de uma classe                   | `meu_carro = Carro("Toyota", "Corolla")` |
| **Atributo**     | Variável dentro da classe                 | `self.modelo = modelo`                   |
| **Método**       | Função dentro da classe                   | `def ligar(self):`                       |
| **Herança**      | Classe filha reutiliza código da classe pai | `class Cachorro(Animal):`               |
| **Polimorfismo** | Métodos com o mesmo nome em classes diferentes | `def voar(self):`                   |
